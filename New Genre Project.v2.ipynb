```python
df = data['Genres'].str.get_dummies(',').stack().sum(level=1).sort_values(ascending = False)
df
```

```python
count = pd.DataFrame(data = df, columns = ['Count']).reset_index()
count.rename(columns={'index':'Genre'},inplace=True)
count
```
